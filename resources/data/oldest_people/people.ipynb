```python
import pandas as pd

people = pd.read_csv('./centenarians.csv')
# https://clio-infra.eu/Indicators/LifeExpectancyatBirthTotal.html
expectancy = pd.read_excel('./life_expectancy.xlsx', index_col=0, header=[0])

```

```python
expectancy
```

```python
expectancy.head()
expectancy = expectancy.fillna(method='backfill', axis=1)
expectancy = expectancy.fillna(method='ffill', axis=1)
```

```python
people['birth_year'] = people['birth_date'].str.slice(0,4)
```

```python
data = people.join(expectancy.set_index('country name'), on="place_of_death_or_residence", how="left")
```

```python
data['2022']
```

```python
data['birth_exp'] = data.apply(lambda r: r[int(r.birth_year)], axis=1)
data['exp_recent'] = data['2022']
```

```python
subdata = data.loc[:, ['rank', 'name', 'birth_date','death_date',
                       'age', 'place_of_death_or_residence', 'gender', 'still_alive', 'birth_exp','exp_recent']]
subdata
```

```python
with open('./centenarians_with_expectations.csv', 'w') as outfile:
    outfile.write(subdata.to_csv(index=False))
```
