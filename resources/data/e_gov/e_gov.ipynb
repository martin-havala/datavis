```python
import pandas as pd

# https://datacatalog.worldbank.org/search/dataset/0050602/Data-Governance-Maturity-Indicators
data = pd.read_excel('./WBG_GovTech Dataset_Mar2023.xlsx', header=[0],sheet_name='CG_GTMI_Data')
```

```python
data.replace(to_replace=r'^-$', value='', regex=True,inplace=True)
data.replace(to_replace=r'\n', value=' ', regex=True,inplace=True)
data
```

```python
with open('./e_gov.csv', 'w') as outfile:
    outfile.write(data.to_csv(index=False))
```
