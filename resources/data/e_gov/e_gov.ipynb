```python
import pandas as pd

# https://datacatalog.worldbank.org/search/dataset/0050602/Data-Governance-Maturity-Indicators
data = pd.read_excel('./WBG_DGSS-dataset_2021-07-22_EDIT.xlsx', header=[1,2],sheet_name='Sheet 1')
```

```python
data.replace(to_replace=r'^-$', value='', regex=True,inplace=True)
```

```python
sdata =  data.set_axis(data.columns.map('_'.join), axis=1)
sdata
```

```python
with open('./e_gov.csv', 'w') as outfile:
    outfile.write(sdata.to_csv())
```
