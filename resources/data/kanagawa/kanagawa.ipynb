```python
import pandas as pd;

from datetime import datetime


# https://open-meteo.com/en/docs/marine-weather-api#latitude=35.47&longitude=139.63&daily=wave_height_max,wave_direction_dominant&timezone=Europe%2FBerlin&start_date=2022-08-01&end_date=2023-04-19
sea = pd.read_csv('./kanagawa_prediction.csv',  parse_dates=['time'])
# https://www.kaggle.com/datasets/lsind18/full-moon-calendar-1900-2050
moon = pd.read_csv('./moon.csv',sep=';')
```

```python
moon
```

```python
import datetime

moon['date']=moon['Date'].apply(lambda x : datetime.datetime.strptime(x, '%Y %b %d') if x else 0)

moon.drop(columns=['Date','Time'], inplace=True)
```

```python

data = sea.rename(columns={'wave_height_max (m)':'height', 'wave_direction_dominant (°)':'direction'}).join(moon.rename(columns={'Moon Phase':'phase' }).set_index('date'), on ='time')
data.head(20)
```

```python

with open('./kanagawa.json', 'w') as outfile:
    outfile.write(data.to_json(orient="records"))
```
