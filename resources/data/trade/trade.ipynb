```python
import pandas as pd

# https://datacatalog.worldbank.org/search/dataset/0050602/Data-Governance-Maturity-Indicators
trades = pd.read_excel(
    'resourcetradeearth-all-all-all-2020.xlsx', header=[0], sheet_name='Trades')
```

```python
trades.drop(columns=['Exporter M.49', 'Importer M.49', 'Exporter region', 'Exporter', 'Importer',
            'Importer region', 'Embodied land (ha)', 'Embodied blue water (m³)'], inplace=True)

trades.rename(columns={
    'Exporter ISO3': 'Exporter',
    'Importer ISO3': 'Importer',
    'Value (1000USD)': 'Value',
    'Weight (1000kg)': 'Weight'
}, inplace=True)
```

```python
trades[(trades.Value.notna() & (trades.Weight.isna())) |
       (trades.Value.isna() & (trades.Weight.notna()))]
```

```python
trades2020 = trades[trades.Year == 2020].dropna().drop(columns=['Year'])
trades2020
```

```python
grp = trades2020.groupby(['Exporter','Importer']).agg({'Value':sum, 'Weight':sum})
```

```python
countries = pd.Series(grp.index.map(lambda x: x[0]).join(grp.index.map(lambda x: x[1])).unique())
```

```python
with open('./trades2020.csv', 'w') as outfile:
    outfile.write(grp.to_csv(index=True))

with open('./countries2020.csv', 'w') as outfile:
    outfile.write(countries.to_csv(index=True))
```
