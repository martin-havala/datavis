```python
import pandas as pd

# https://population.un.org/wpp/Download/Standard/CSV/
# Population	 1950-2100, all scenarios (ZIP, 10.52 MB)    Total Population on 01 July.

pop = pd.read_csv('WPP2022_TotalPopulationBySex.csv')
```

```python
pop.describe()
```

```python
pop.head()
```

```python
pop.drop(columns=['SortOrder', 'LocID',	'Notes',
         'LocTypeID', 'ISO2_code', 'VarID', 'MidPeriod'], inplace=True)
pop[pop['Time'] == 2100]
```

```python
countries = pop[pop['LocTypeName'].isin(['Country/Area', 'World'])].drop(columns='LocTypeName')
variants = ['Medium', 'High', 'Low']
```

```python
medium = countries[countries['Variant'].isin(variants
    ) & (countries['Time'] % 10 == 0) & (countries['Time'] >= 2020)]
medium
```

```python
tst = medium.set_index(['ISO3_code', 'SDMX_code', 'ParentID', 'Location', 'Time', 'Variant'], drop=True) \
    .unstack(['Variant','Time'])
# .pivot(index=['ISO3_code', 'SDMX_code', 'LocTypeName', 'ParentID', 'Location', 'Time'], columns=['Variant'], values=['MidPeriod', 'PopMale', 'PopFemale', 'PopTotal', 'PopDensity'])  # .unstack('Variant')
tst
```

```python
tst.columns = ['_'.join((col[0], str(col[1]), str(col[2])))
               for col in tst.columns]
categories = ['Male', 'Female', 'Total', 'Density']
tst
```

```python

for c in categories:
    for v in variants:
        selection = 'Pop' + c + '_' + v
        selection2030 = selection + '_2030'
        selection2040 = selection + '_2040'
        selection2050 = selection + '_2050'
        selection2060 = selection + '_2060'
        selection2070 = selection + '_2070'
        selection2080 = selection + '_2080'
        selection2090 = selection + '_2090'
        selection2100 = selection + '_2100'
        tst[selection] = tst.apply(lambda x: list([
            x[selection2030],
            x[selection2040],
            x[selection2050],
            x[selection2060],
            x[selection2070],
            x[selection2080],
            x[selection2090],
            x[selection2100],
        ]), axis=1)

for c in categories:
    selection = 'Pop' + c + '_Medium'
    selection2020 = selection + '_2020'
    tst['Pop'+c+'_2020'] = tst[selection2020]

tst
```

```python
tst.columns[100]
```

```python
res = tst.loc[:, ['PopMale_2020', 'PopFemale_2020', 'PopTotal_2020', 'PopDensity_2020','PopMale_Medium', 'PopMale_High', 'PopMale_Low',
                  'PopFemale_Medium', 'PopFemale_High', 'PopFemale_Low', 'PopTotal_Medium', 'PopTotal_High', 'PopTotal_Low', 'PopDensity_Medium', 'PopDensity_High', 'PopDensity_Low']]
res
```

```python
with open('./population.json', 'w') as outfile:
    outfile.write(res.reset_index().to_json(orient="records"))
```
