```python
name_basics = pd.read_csv("./name.basics.tsv", sep='\t', )
title_crew = pd.read_csv("./title.crew.tsv", sep='\t')
title_basics = pd.read_csv("./title.basics.tsv", sep='\t', low_memory=False)
```

```python
cooperations= title_crew[title_crew['directors'].str.contains(',') ]
```

```python
# movies = title_basics.loc[title_basics['titleType'].isin(['short', 'movie', 'tvShort', 'tvMovie'])]
movies = title_basics.loc[title_basics['titleType'].isin(['movie'])]
movies
```

```python
hlp = cooperations.set_index('tconst').join(movies.set_index('tconst'),how='inner',)
hlp
cooperations = hlp.reset_index().loc[:,['tconst','directors']]
```

```python
cooperations
```

```python

cooperations['dir_list']= cooperations['directors'].apply(lambda x: (x.split(',')))
```

```python
cooperations
```

```python
cooperations.sort_values(by="dir_count")
```

```python
movie_directors  = cooperations.explode('dir_list')
movie_directors
```

```python
movie_count = movie_directors.groupby("dir_list")["tconst"].nunique()
movie_count
```

```python
d1 = movie_directors[['tconst','dir_list']]
d2 = movie_directors[['tconst','dir_list']]

d1
```

```python
links = d1.set_index('tconst').join(d2.set_index('tconst'), on="tconst", lsuffix='1', rsuffix='2')
links
```

```python
links = links[links['dir_list1']!=links['dir_list2']]
```

```python
links.reset_index(inplace=True)
links
```

```python
links[links['dir_list1']=='nm0000041']
```

```python
all_nodes  = links[links['dir_list1']!=links['dir_list2']].groupby(['dir_list1']).agg({"tconst": pd.Series.nunique })
# all_nodes['tconst']= all_nodes['tconst'].apply(lambda x: x/2) 
```

```python
treshold = 5
final_nodes = all_nodes[all_nodes['tconst'] >= treshold ]
```

```python
hlp = links[(links['dir_list1']!=links['dir_list2'])  & (links['dir_list1'].isin(final_nodes.index.to_series())) & (links['dir_list2'].isin(final_nodes.index.to_series()))
].set_index('dir_list1')

final_links = hlp.groupby(['dir_list1', 'dir_list2']).agg({"tconst": pd.Series.nunique})
```

```python
c = links.reset_index()
c[c['dir_list1']=='nm9110144']
```

```python
nodes_json = final_nodes.reset_index().rename(columns={'dir_list1':'id','tconst':'count'}).to_json(orient='records')

nodes_json[0:500]
```

```python
a = final_links.reset_index()
a[a['dir_list1']=='nm9110144']
```

```python

json_links = final_links.reset_index().rename(columns={'dir_list1':'source', 'dir_list2':'target', 'tconst':'value'}).to_json(orient='records')

json_links[0:500]
```

```python
crew = name_basics.loc[:,['nconst','primaryName']]
final_directors = final_nodes.join(crew.set_index('nconst'), how="inner").rename(columns={'tconst':'count'})
```

```python
with open('./network/links.json', 'w') as outfile:
    outfile.write(json_links)

with open('./network/nodes.json', 'w') as outfile:
    outfile.write(nodes_json)

with open('./network/directors.json', 'w') as outfile:
    outfile.write(final_directors.to_json(orient="index"))
```
